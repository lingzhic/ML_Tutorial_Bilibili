# Random Gradient Descent

```python
import dataset
from matplotlib import pyplot as plt
import numpy as np

xs, ys = dataset.get_beans(100)
```

```python
# randomly assign w and b value, bad fitting
w = 0.1
b = 0.1
y_pre = w * xs + b

plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
# Plot a dynamic plot
from IPython import display

alpha = 0.01   # step factor
for m in range(500):
    for i in range(100):
        x = xs[i]
        y = ys[i]
        """
        Since cost function e = (y - (w*x + b))^2   -> 此处的y和x是dataset上的某一个单独的点
        针对w来说：
        e = 2xw + x2w2 
        e = b^2 + (2*x*w - 2*y)*b + (x^2*w^2 + y^2 - 2*x*y*w)
        """
        dw = 2 * x**2 * w + 2*x*b - 2*x*y    # 对 w 求导
        db = 2*b + 2*x*w - 2*y               # 对 b 求导
        w = w - alpha * dw                   # update w 值 according to current gradient
        b = b - alpha * db                   # update b 值 according to current gradient
        
    # plot a dynamic fitting curve
    plt.clf()
    plt.scatter(xs,ys)
    y_pre = w * xs + b
    plt.xlim(0, 1)
    plt.ylim(0, 1.3)
    plt.plot(xs, y_pre)
    display.clear_output(wait=True)
    plt.pause(0.01)
```
