# Cost function
error as a function of w (slope of prediction function y = w*x)

```python
import dataset
from matplotlib import pyplot as plt
import numpy as np

xs, ys = dataset.get_beans(100)
```

```python
# randomly assign a w value
w = 0.1
y_pre = w * xs

plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
# an error value at w = 0.1 (value of e(w=0.1))
e = (ys - y_pre)**2
sum_e = np.sum(e)/len(e)
print(sum_e)
```

Cost function (for prediction function in the form of y = kx) is

$error = \frac{1}{m} \sum\limits_{i=0}^{m} (y_i - y_{pre})^2 = \frac{1}{m} \sum\limits_{i=0}^{m} (y_i - x_i w)^2$

$error = \frac{1}{m} \sum\limits_{i=0}^{m} (x_i^2 w^2 + (-2x_i y_i)w + y_i^2)$

```python
# plot the e value at different w value (a parabolic curve)
# show the plot of the cost function
ws = np.arange(0, 3, 0.1)

# a for loop to get error value at different w value, append to list es
es = []
for w in ws:
    y_pre = w * xs
    # note that the real form of this e equation is e = (ys - w*xs)**2 = ys**2 - 2*xs*ys*w + xs**2 * w**2, i.e. a = xs**2, b = -2*xs*ys
    e = (ys - y_pre)**2
    sum_e = np.sum(e)/len(e)
    es.append(sum_e)

plt.title("cost function", fontsize=12)
plt.xlabel("w")
plt.ylabel("e")
plt.plot(ws, es)
plt.show()
```

Get the bottom of the parabolic curve using equation: $ min = -\frac{b}{2a} $

It shows a perfect fitting!!!

# Normal equation
get the bottom of the above parabolic curve using equation: $min = -\frac{b}{2a}$

```python
# Normal equation是一步到位的方法，一次性带入全部的样本（取average得到整体cost function)进行计算, 计算资源吃不消
w_min = - np.sum(-2*xs*ys) / (2 * np.sum(xs*xs))
print(f"The w value when e reaches minimum value is: {w_min}")

# sub w_min value back to the y_pre equation, it shows a perfect fitting!!!
w = w_min
y_pre = w * xs

plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```
