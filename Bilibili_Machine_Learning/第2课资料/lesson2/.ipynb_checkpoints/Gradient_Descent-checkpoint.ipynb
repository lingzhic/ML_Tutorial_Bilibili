# Gradient Descent

```python
import dataset
from matplotlib import pyplot as plt
import numpy as np

xs, ys = dataset.get_beans(100)
```

```python
# randomly assign a w value, bad fitting
w = 0.1
y_pre = w * xs

plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
alpha = 0.1   # step factor
for m in range(200):
    for i in range(100):
        x = xs[i]
        y = ys[i]
        """
        Since cost function e = ys**2 - 2*xs*ys*w + xs**2 * w**2
        a = xs**2
        b = -2*xs*ys
        c = y**2
        gradient
        k = 2*a*w + b
        """
        k = 2 * x**2 * w + (-2*x*y)   # 一个点一个点的修正 w
        w = w - alpha * k


y_pre = w * xs

plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
# Plot a dynamic plot
from IPython import display

alpha = 0.1   # step factor
for m in range(150):
    for i in range(100):
        x = xs[i]
        y = ys[i]
        """
        Since cost function e = ys**2 - 2*xs*ys*w + xs**2 * w**2
        a = xs**2
        b = -2*xs*ys
        c = y**2
        gradient
        k = 2*a*w + b
        """
        k = 2 * x**2 * w + (-2*x*y)
        w = w - alpha * k
        
        plt.clf()
        plt.scatter(xs,ys)
        y_pre = w * xs
        plt.xlim(0, 1)
        plt.ylim(0, 1.2)
        plt.plot(xs, y_pre)
        display.clear_output(wait=True)
        plt.pause(0.01)
```
