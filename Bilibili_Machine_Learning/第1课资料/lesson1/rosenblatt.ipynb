```python
import dataset
from matplotlib import pyplot as plt
```

```python
xs,ys = dataset.get_beans(100)

print(xs)
print(ys)
```

```python
# configure figure settings
plt.title("Size-Toxicity function", fontsize=12)
plt.xlabel("Bean size")
plt.ylabel("Toxicity")

plt.scatter(xs,ys)
plt.show()
```

```python
# set function y = 0.5x (initialise the factor as 0.5)
w = 0.5
```

```python
'''
# a single update of w
x = xs[0]
y = ys[0]
y_pre = w * x
err = y - y_pre
alpha = 0.05    # adjusting factor
w = w + alpha * err * x
plt.show()
'''
```

```python
# a single loop adjustment of w, not good enough
alpha = 0.05  # adjusting factor
for i in range(100):
    x = xs[i]
    y = ys[i]
    y_pre = w * x
    err = y - y_pre
    w = w + alpha * err * xs    # alpha -> adjusting step   err -> error between prediction and training value   xs -> current x value (used to ensure the adjustment is in correct direction)

y_pre = w * xs
plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```

```python
# repeat w adjusting for 100 times, good enough
alpha = 0.05  # adjusting factor
for m in range(100):
    for i in range(100):
        x = xs[i]
        y = ys[i]
        y_pre = w * x
        err = y - y_pre
        w = w + alpha * err * x

y_pre = w * xs
plt.scatter(xs,ys)
plt.plot(xs, y_pre)
plt.show()
```
