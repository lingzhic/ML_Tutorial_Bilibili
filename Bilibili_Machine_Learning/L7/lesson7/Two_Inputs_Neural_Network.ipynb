## Generate 3D data

```python
import numpy as np
import dataset
import plot_utils

# Generate the dataset
SAMPLE_SIZE = 100
xs,ys = dataset.get_beans(SAMPLE_SIZE)
plot_utils.show_scatter(xs, ys)
```

## Create a two inputs neuro

<img src="Two_Inputs_neuro.png" alt="A simple neural network" width="200"/>

```python
w1 = 0.1
w2 = 0.1
b = 0.1

x1s = xs[:,0]
x2s = xs[:,1]

def sigmoid(z):
    return 1/(1+np.exp(-z))

def forward_propagation(x1s, x2s):
    z = w1 * x1s + w2 * x2s + b
    a = sigmoid(z)
    return a

# Pass xs, ys, and forward_propagation function to this scatter surface plotting function
plot_utils.show_scatter_surface(xs, ys, forward_propagation)
```

## Machine learning process

```python
# %matplotlib tk
for _ in range(5000):
    for i in range(SAMPLE_SIZE):
        x = xs[i]           # Note that x is an array but not a single number
        y = ys[i]

        x1 = x[0]
        x2 = x[1]

        # Forward propagation
        a = forward_propagation(x1, x2)

        # get the error
        e = (y - a)**2

        # get the gradient for prediction function
        # z = w1 * x1 + w2 * x2 + b
        deda = -2 * (y - a)
        dadz = a * (1 - a)
        dzdw1 = x1
        dzdw2 = x2
        dzdb = 1

        dedw1 = deda * dadz * dzdw1
        dedw2 = deda * dadz * dzdw2
        dedb = deda * dadz * dzdb

        # Update the w1, w2, and b value to finish backward propagation
        alpha = 0.01
        w1 = w1 - alpha * dedw1
        w2 = w2 - alpha * dedw2
        b = b - alpha * dedb

# Finally plot the curve
plot_utils.show_scatter_surface(xs, ys, forward_propagation)

```
