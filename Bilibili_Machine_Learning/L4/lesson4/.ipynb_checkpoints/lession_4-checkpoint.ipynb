```python
import matplotlib.pyplot as plt
import dataset
import numpy as np
```

```python
#从数据中获取随机豆豆
m=100
xs,ys = dataset.get_beans(m)
```

```python
#配置图像
plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel('Bean Size')
plt.ylabel('Toxicity')
plt.xlim(0, 1)
plt.ylim(0, 1.5)

# 豆豆毒性散点图
plt.scatter(xs, ys)
```

```python
#预测函数
w=0.1
b=0.1
y_pre = w*xs+b

plt.plot(xs, y_pre)
```

```python
ws = np.arange(-1, 2, 0.1)

es = []
for w in ws:
    y_pre = w * xs + b
    e = np.sum((ys - y_pre)**2) * (1/m)  # m is the number of samples, already defined before
    es.append(e)

plt.plot(ws, es)    # 此处的图像是当b = 0.1的曲线
    
```

```python
# 现在来画b和w都变的3D 图像
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.set_zlim(0, 2)   # Set z axis range limitation to get a more prominent curvature


ws = np.arange(-1, 2, 0.1)
bs = np.arange(-2, 2, 0.01)

for b in bs:
    es = []
    for w in ws:
        y_pre = w * xs + b
        e = np.sum((ys - y_pre)**2) * (1/m)  # m is the number of samples, already defined before
        es.append(e)

    ax.plot(ws, es, b, zdir='y')    # 设置z轴垂直朝上，绘制一条w变化而b不变的曲线
# Get multiple curves with different w and b

plt.show()
```
